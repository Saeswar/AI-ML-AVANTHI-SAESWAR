```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [1,2,3]
y = [1,2,3]
plt.plot(x,y)
plt.figure(figsize=(5,3),dpi=100)
```

```python
x=np.arange(0,5,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title('our first sin graph',fontdict ={'fontsize':35})
```

```python
x = [2,4,6,8,10,12]
y = [3,6,9,12,15,18]
plt.plot(x,y)
plt.title('my first graph',fontdict ={'fontsize':10})
plt.xlabel('x cordinates')
plt.ylabel('y cordinates')
plt.xticks([2,4,6,8,10,12])
plt.yticks([3,6,9,12,15,18])
```
