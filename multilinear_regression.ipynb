```python
df = pd.read_csv('/content/1000_Companies.csv')
df
```

```python
X = df.drop('Profit',axis = 'columns')
Y = df['Profit']
```

```python
X
```

```python
Y
```

4.find correlations using heatmap

```python
sns.heatmap(df.corr())
```

5.identify categorical Data and Encode it

```python
dummies = pd.get_dummies(df['State'])
dummies
```

```python
merged = pd.concat([X,dummies], axis = 'columns')
merged
```

```python
X = merged.drop(['State','California'],axis = 'columns')
X
```

6.split the data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2)
```

7.train your model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```

8.predict values

```python
 Y_pred = model.predict(X_test)
 Y_pred
```

```python
Y_test
```

9.model metrcis calculation

```python
model.score(X_test,Y_test)
```

```python
dummies1 = pd.get_dummies(df['fuel'])
dummies2 = pd.get_dummies(df['seller_type'])
dummies3 = pd.get_dummies(df['transmission'])
```

```python
merged = pd.concat([X,dummies1,dummies2,dummies3], axis = 'columns')
merged
```

```python
X = merged.drop(['transmission','fuel','CNG','Petrol','Electric','LPG','name','seller_type','transmission'],axis = 'columns')
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
```
