```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

plt.grid(color = 'violet', linestyle = '--', linewidth = 0.5)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4,5])
y = np.array([1, 2, 3, 4,5])
font1 = {'family':'serif','color':'blue','size':10}
font2 = {'family':'serif','color': 'red','size':15}
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title('first graph',fontdict=font1)
x = np.array([1, 2, 3, 4,5])
y = np.array([5, 4, 3, 2,1])

plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.title('second graph',fontdict=font1)
x = np.array([1, 2, 3, 4,5])
y = np.array([5, 4, 3, 2,1])

plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.title('third graph',fontdict=font1)
x = np.array([5, 4, 3, 2,1])
y = np.array([5, 4, 3, 2,1])

plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.title('fourth graph',fontdict=font1)
plt.subplots_adjust(wspace=0.4,hspace=0.4)
plt.suptitle('all graphs',fontdict=font2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 1, 1, 1,1])
y = np.array([1, 2, 3, 4,5])
plt.scatter(x, y,color = '#A52A2A')


x = np.array([1,2,3,4,5])
y = np.array([3,3,3,3,3])
plt.scatter(x, y,color = '#8B008B')


x = np.array([6,6,6,6,6])
y = np.array([1,2,3,4,5])
plt.scatter(x, y,color = '#B8860B')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
